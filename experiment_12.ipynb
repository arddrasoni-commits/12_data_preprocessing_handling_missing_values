```python
import pandas as pd

data = {
    "Roll no.": [23,34,45,56,67],
    "Name": ["Slash","Shadow","Charlie","Whistle", "Tango"],
    "Age": [19,"-", 23,24,13],
    "Marks": [85, 90, "-",55,56],
    "Department":["Acoustics","Aerodynamics","ENTC","Navy Tech","Mastering"],
    "Admission_Date":["01-09-20", "02-04-23", "04-06-24", "06-12-12", "08-10-24"]
}

df=pd.DataFrame(data)
df
```

```python
df.to_csv("student_data.csv",index = np.False_)
```

```python
# Step 2: Replace '-' with NaN to mar missing values
df.replace("-",np.nan,inplace= True)

#convert columns to numeric where needed

df["Age"]= pd.to_numeric(df["Age"], errors = 'coerce')
df["Marks"]= pd.to_numeric(df["Marks"], errors = 'coerce')

''' when converting data like strings to datatime, if some values are invalid or cannot be converted
then instead of showing an error, it will replace these invalid values with NaT(Not a Time)'''
df
```

```python
#Step 3: Display missing values
print("Missing Values")
print(df.isnull().sum()) #identifies all null/NaN values and sums it
```

```python
# Fill missing Age with mean
df["Age"].fillna(df["Age"].mean(), inplace= True)

# Fill missing Marks with median
df["Marks"].fillna(df["Marks"].median(), inplace= True)


```

```python
# Step 5: Fix inconsistent text formatting

df["Department"] = df["Department"].str.upper()
```

```python
# Step 6: convert admission_date to standard date time format
df["Admission_date"] = pd.to_datetime(df["Admission_Date"], errors = 'coerce')
```

```python
# Step 7: Cleaned Data
print("Cleaned Data")
df
```

```python
# Step 8: Save cleaned dataset
df.to_csv("cleaned_student_data.csv",index = False)

print("\nCleaned csv file saved as cleaned_student_data_.csv")
```

```python
import pandas as pd

NewData = {
    "Fruits" : ["Apple","Banana","Mango","Grapes"],
    "Calories" : [420,380,"_",210],
    "Weight" : ["_",40,70,3],
    "Taste" : ["Sweet","Sweet","Sour","Sour"],
    "Price" : [70,40,"_",100],
    "DateOfPurchace" : ["2023-03-12","2020-01-06","03-01-2025","2026/02/12"]
}

df = pd.DataFrame(NewData)
print("Original DataFrame:")
df.to_csv("Fruits.csv",index=False)
print(df)
```

```python
import numpy as np

df.replace("_",np.nan,inplace=True)
print(df)
```

```python
df["Calories"]=pd.to_numeric(df["Calories"],errors="coerce")
df["Weight"]=pd.to_numeric(df["Weight"],errors="coerce")
df["Price"]=pd.to_numeric(df["Price"],errors="coerce")
print(df)
```

```python
print(df.isnull().sum())
```

```python
print(df["Calories"].fillna(df["Calories"].mean(),inplace=True))
print(df["Weight"].fillna(df["Weight"].median(),inplace=True))
print(df["Price"].fillna(df["Price"].mean(),inplace=True))
```

```python
df["Fruits"] = df["Fruits"].str.upper()
df["Taste"] = df["Taste"].str.lower()
df
```

```python
print("Cleaned data : \n",df)
df.to_csv("CleanedData.csv",index=False)
```
